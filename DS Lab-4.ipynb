```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('Titanic_Dataset.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
# finding missing categorical values
feature_nan = [features for features in dataset.columns 
               if dataset[features].isnull().sum()>1 and 
                dataset[features].dtype=='O']
#if datatype is 'O' or 'o' then it is considered as categorical
for feature in feature_nan:
    print(feature, np.round(dataset[feature].isnull().mean(),3), 
          '% missing value')
```

```python
# replacing categorical missing values
def replace_cat_features(dataset, feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('Missing') 
    # replacing the missing values with label 'Missing'
    return data

dataset=replace_cat_features(dataset,feature_nan)
dataset.head()
```

```python
dataset[feature_nan].isnull().sum()
```

```python
# finding missing numerical values
numerical_with_nan = [features for features in dataset.columns 
               if dataset[features].isnull().sum()>1 and 
                dataset[features].dtype!='O']
for feature in numerical_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(),3), 
          '% missing value')
```

```python
# replacing the numerical missing values
for i in numerical_with_nan:
    median_value = dataset[i].median()
    dataset[i + ' nan'] = np.where(dataset[i].isnull(),1,0)
    dataset[i].fillna(median_value,inplace=True)
dataset[numerical_with_nan].isnull().sum()
```

A new column 'Age nan' was created to capture the nan values

```python
dataset.head()
```

```python
num_features=['Age','Fare']
for feature in num_features:
    dataset[feature]=np.log(dataset[feature],where=dataset[feature] > 0)
```

```python
dataset.head()
```

```python
categorical_features=[feature for feature in dataset.columns 
                      if dataset[feature].dtype=='O']
```

```python
categorical_features
```

```python
# finding rare categorical variables
for feature in categorical_features:
    temp=dataset.groupby(feature)['Fare'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),
                              dataset[feature],'Rare_var')
```

```python
dataset.head()
```

We can see that columns 'Name' and 'Ticket' are present less than 1% of the observation

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['Fare'].mean().
    sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

### CONCLUSION:
Feature Engineering was performed on the 'Titanic' dataset. We found that columns 'Cabin' and 'Embarked' are the categorical variables that had missing values, the missing values were replaced by the label 'Missing'. Numerical column 'Age' had some missing values which were replaced by using median. Since the numerical variables are skewed, log normal distribution was performed. Categorical variables 'Name' and 'Ticket' are present less than 1% of the observation.
